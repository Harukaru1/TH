1. Sales

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales.csv")
```

```python
df
```

✓ Fix column datatypes

```python
df['amount'] = df['price'].str.replace('$', '').astype(float)
df.head(10)
```

```python
df = df.drop('price', axis=1)
```

```python
df.head(10)
```

```python
moving_column = df.pop('amount')
df.insert(3, 'amount', moving_column)
df.head(10)
```

```python
df = df.rename(columns={'amount': 'price'})
df.head(10)
```

```python
df['amount'] = df['line_total'].str.replace('$', '').astype(float)
df.head(10)
```

```python
df = df.drop('line_total', axis=1)
```

```python
df = df.rename(columns={'amount': 'line_total'})
df.head(10)
```

✓ Drop if duplicated or null.

```python
df = df.dropna(how='all')
```

```python
df = df.drop_duplicates(keep='last')
```

```python
df
```

✓ Sanity check for value ranges and to check assumptions.

```python
if any(df['price']<0):
    print('Error: values outside expected range')
```

```python
df = df[df['price'] >- 0]
```
